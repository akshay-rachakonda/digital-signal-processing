```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simplest
a= np.arange(0,10,1)

plt.plot(a)
plt.show()
```

```python
a= np.arange(0,10,1)
x= 0.1*a # create x axis 
plt.plot(x,a)
plt.show()
```

```python
a= np.arange(0,10,1)
x= 0.1*a # create x axis
b= 2*a
plt.plot(x,a,x,b)# plot 2 wave form same figure
plt.show()
```

```python
# sine cosine
f= 50
omega= 2*np.pi*f
t_duration=1
fg=1e+6
Tg=1/fg
t_array= np.arange(0,t_duration,Tg)
sin_array= np.sin(omega*t_array)
cos_array= np.cos(omega*t_array)
plt.plot(t_array,sin_array)
```

```python
plt.plot(t_array,sin_array) # limit x axis to plot
plt.plot(t_array,cos_array)
plt.xlim([0, 0.04])
```

```python
# labels
plt.plot(t_array,sin_array,label='sin') # limit x axis to plot
plt.plot(t_array,cos_array,label='cos')
plt.legend()
plt.xlim([0, 0.04])
```
