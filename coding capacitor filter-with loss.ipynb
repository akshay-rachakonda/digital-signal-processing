```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# signal Current as input 
Imag= 5
f= 50
omega= 2*np.pi*f
Tg= 1e-6 #microseconds
t= np.arange(0,1,Tg)
i= Imag*np.sin(omega*t)
```

```python
plt.plot(t,i,ds="steps")
plt.xlim([0,0.02])
plt.show()
```

```python
# sampling at 5000 Hz
# original signal develops at 1 micro sec, if choose 1 sample at 1/5000 sec i.e. 200 micro sec
# 200/1= 200 we are taking only 1 in 200 samples generated
# it runs for 1 sec so 1000000/200 samples willbe there= 5000
T=1/5000
n_samples=int(T/Tg)
t_samples= t[::n_samples]
i_samples=i[::n_samples]
plt.plot(t_samples,i_samples,ds="steps") # matplotlib.plt does not interpolate between samples
plt.xlim([0,0.02])
plt.show()
```

```python
# looping through numpy array
for i in np.nditer(i_samples): # returns value of the samples
    print(i)
```

```python
# looping through array
for i in np.ndenumerate(i_samples):#np.ndenumerate returns tuple of index and sample value
    print(i)
```

```python
# begin calculation
# for both input and output we need two samples present and past values
#input
u= np.zeros(2)
#output
y=np.zeros(2)
#voltage samples
v_samples=np.zeros(i_samples.size)
# for loop
C=10e-6
R= 2000
for i_index, i_val in np.ndenumerate(i_samples):
    u[0]=i_val # present value
    y[0]=(T*u[0]+T*u[1]+2*C*y[1]-T*y[1]/R)/(2*C+T/R)
    u[1]=u[0] #past value
    y[1]=y[0]
    v_samples[i_index]=y[0]
```

```python
plt.figure()
plt.plot(t_samples, i_samples, label="input")
plt.plot(t_samples, v_samples, label="output",ds="steps")
plt.xlim([0,0.08])
plt.title("Input and output")
plt.legend()
plt.show()
```
